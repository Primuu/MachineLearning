## Library Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

## Dataset Import

```python
data = pd.read_csv('Kruk_VLagun_156.csv')
data.shape
```

## Dataset properties

```python
data.head()
```

```python
data.describe()
```

## Visualization 

```python
data.plot(x='SS', y='SecchDisc', style='o')
plt.title('SS vs SecchDisc')
plt.xlabel('SS')
plt.ylabel('SecchDisc')
plt.show()
```

## Distribution

```python
plt.figure(figsize=(15, 10))
plt.tight_layout()
sns.distplot(data['SecchDisc'])
```

## Dividing data to attributes (x) and labels (y)

```python
X = data['SS'].values.reshape(-1, 1)
y = data['SecchDisc'].values.reshape(-1, 1)
```

## Splitting data to train and test sets 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Model training

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Viewing intercept and regressor coefficient values

```python
regressor.intercept_
```

```python
regressor.coef_
```

## Predicting using trained model

```python
y_pred = regressor.predict(X_test)
```

## Comprising real data with predicted values 

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

## Comparison visualization

```python
df1 = df.head(25)  # limiting the number of entries
df1.plot(kind='bar', figsize=(16, 10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
plt.scatter(X_test, y_test, color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

#### Model works fine

## Error Measures

```python
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

RMSE equals ~0.084 which is ~22% of mean value (0.376026) - model isn't perfect, but it can still make pretty good predictions
