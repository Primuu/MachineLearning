## Library Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
np.random.seed(seed=13)
```

## Generating dataset

```python
x = np.random.uniform(0, 1, 20)
```

## Function definition

```python
def f(x):
    return x * 2
```

## Creating target with noise

```python
noise_variance = 0.2
noise = np.random.randn(x.shape[0]) * noise_variance
```

```python
t = f(x) + noise
```

## Visualization

```python
plt.figure(figsize=(5, 3))
plt.plot(x, t, 'o', label='$t$')
plt.plot([0, 1], [f(0), f(1)], 'b--', label='$f(x)$')
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$t$', fontsize=12)
plt.axis((0, 1, 0, 2))
plt.title('inputs (X) vs targets (t)')
plt.legend(loc=2)
plt.show()
```

## Neural network model and loss function definitions

```python
def nn(x, w):
    """Output function y = x * w"""
    return x * w
```

```python
def loss(y, t):
    """MSE loss function"""
    return np.mean((t - y) ** 2)
```

## Loss function visualization (given weight w)

```python
ws = np.linspace(0, 4, num=100)
loss_ws = np.vectorize(lambda w: loss(nn(x, w), t))(ws)
```

```python
plt.figure(figsize=(5, 3))
plt.plot(ws, loss_ws, 'r--', label='loss')
plt.xlabel('$w$', fontsize=12)
plt.ylabel('$\\xi$', fontsize=12)
plt.axis((0, 1, 0, 2))
plt.title('loss function with respect to $w$')
plt.xlim(0, 4)
plt.legend()
plt.show()
```

## Gradient function

```python
def gradient(w, x, t):
    """Gradient function. (Remember that y = nn(x, w) = x * w)"""
    return 2 * x * (nn(x, w) - t)
```

## Delta_w function

```python
def delta_w(w_k, x, t, learning_rate):
    """Update function delta w"""
    return learning_rate * np.mean(gradient(w_k, x, t))
```

## Initial parameters

```python
w = np.random.rand()
learning_rate = 0.9
```

## Gradient loop

```python
nb_of_iterations = 4
w_loss = [(w, loss(nn(x, w), t))]
for i in range(nb_of_iterations):
    dw = delta_w(w, x, t, learning_rate)
    w = w - dw
    w_loss.append((w, loss(nn(x, w), t)))
```

## Results

```python
for i in range(0, len(w_loss)):
    print(f'w({i}): {w_loss[i][0]:.4f} \t loss: {w_loss[i][1]:.4f}')
```

## #2 Gradient loop

```python
w = np.random.rand()
learning_rate = 0.1

nb_of_iterations = 40
w_loss_2 = [(w, loss(nn(x, w), t))]
for i in range(nb_of_iterations):
    dw = delta_w(w, x, t, learning_rate)
    w = w - dw
    w_loss_2.append((w, loss(nn(x, w), t)))
    
for i in range(0, len(w_loss_2)):
    print(f'w({i}): {w_loss_2[i][0]:.4f} \t loss: {w_loss_2[i][1]:.4f}')
```

## #3 Gradient loop

```python
w = np.random.rand()
learning_rate = 0.8

nb_of_iterations = 7
w_loss_3 = [(w, loss(nn(x, w), t))]
for i in range(nb_of_iterations):
    dw = delta_w(w, x, t, learning_rate)
    w = w - dw
    w_loss_3.append((w, loss(nn(x, w), t)))
    
for i in range(0, len(w_loss_3)):
    print(f'w({i}): {w_loss_3[i][0]:.4f} \t loss: {w_loss_3[i][1]:.4f}')
```

## The gradient descent updates visualization

```python
plt.figure(figsize=(6, 4))
plt.plot(ws, loss_ws, 'r--', label='loss')

for i in range(0, len(w_loss)-1):
    w1, c1 = w_loss[i]
    w2, c2 = w_loss[i+1]
    plt.plot(w1, c1, 'bo')
    plt.plot([w1, w2], [c1, c2], 'b-')
    plt.text(w1, c1+0.05, f'${i}$')
plt.plot(w2, c2, 'bo', label=f'$w(k)$')
plt.text(w2, c2+0.05, f'${i+1}$')

plt.xlabel('$w$', fontsize=12)
plt.ylabel('$\\xi$', fontsize=12)
plt.title('Gradient descent updates plotted on loss function')
plt.xlim(0, 4)
plt.legend(loc=1)
plt.show()
```
